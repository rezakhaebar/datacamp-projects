| `description` | Description of the show |
| `genre` | Show genre |

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Overview 10 data
netflix_df.head(10)
```

```python
# Check dataset types
netflix_df.info()
```

```python
# Check summary stats
netflix_df.describe()
```

```python
# Check missing value
netflix_df.isnull().sum()
```

```python
# Movies released in 1990s
movies90 = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] < 2000)]
print(movies90)
```

```python
movies90.head()
```

```python
# Plot the distribution of movie durations
plt.figure(figsize=(10, 6))
plt.hist(movies90['duration'], bins=30, edgecolor='black')
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Most frequent movie duration in 1990
duration = movies90['duration'].mode()[0]
print('The most frequent movie duration: ' + str(duration) + ' minutes')
```

```python
short_movie_count = len(movies90[(movies90['duration'] < 90) & (movies90['genre'] == 'Action')])
print(short_movie_count)
```
